```python
import matplotlib.pyplot as plt
import numpy as np

points = np.array([[1,2],[2,1]])# 在此处修改坐标
t = np.linspace(0,1,1000)
t = np.array([t,t]).T
B1 = (1-t)*points[0]+t*points[1]

plt.plot(B1[:,0],B1[:,1])
plt.plot(points[:,0],points[:,1],'r.')
plt.show()
```
